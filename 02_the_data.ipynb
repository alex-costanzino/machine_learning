# **The data**
Data are never perfect, there may be missing values, inconsistencies, duplicates or wrong data.
*Different types*: quantitative, qualitative, structured, etc...


**Outliers**: small amount of data different from the rest, due to anomalies (interesting) or errors and noise.

## Some glossary
**Discrete domains**: also know as labels, allow a finite or infinitely countable number of values. Nominals and ordinals (possibly binary).

**Continuous domains**: floating point variables. Interval and ratios.

**Asymmetric attributes**: only presence is important (flag value), in particular in discovery of association rules.

## Some general characteristic
**Dimensionality**: large or tall data versus small ones.

**Sparsity**: when there are many zeros or nulls.

**Resolution**
* Too detailed data can be affected by noise;
* Too general data can hide interesting patterns.

Note: there's a widespread bad habit to store zero or some special value when a piece of information is not available.

## Record data
There are different method to store information.

### Relational table
The set of attributes is the same for all the records.

![](https://cdn.hswstatic.com/gif/relational-database-chart.jpg)

### Data matrix
A constrained relation table with numeric attributes.
Each row is a point in a vector space ($N$ rows and $D$ dimensions).
We can use algebra operations.

### Document representation
Typical of text info:
* Each row represents a document;
* Each column represents a term;
* Each cell contains the absolute frequency of the term in the document. The sequence of terms is lost.

### Transactional data
Each record contains a set of objects where the lenght of a row can change (no same numbers of attributes).

### Others
* Graph data;
* Ordered data.

## Data quality

### Noise
It's a (random) modification of original values: mix of interesting data and uninteresting data

### Outliers
Data whose characteristics are considerably different from most of the data in the dataset.
Can be generated by noise or rare events/processes (useful informations).

### Missing values
Happens where data were not collected (i.e. reluctancy in communicate income) or the information is not applicable (i.e. child doesn't have income).

There are different ways to manage them:
* Ignoring the entire instance (row);
* Estimate or put default values (if possible, but lowers the quality of data);
* Ignore (if possible);
* Insert all the possible values weighted with probabilities.

### Duplicated data
Data objects that are duplicates or almost duplicates (i.e. when data are merged from different sources), causes distortion of the data.
Requires data cleaning (difficult process).
